```python
import csv
f = open('./data/gender.csv')
data = csv.reader(f)
m = []
f = []

name = input('이름을 입력하세요')
for row in data :
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break

import matplotlib.pyplot as plt
plt.title(f'{name} 남여인구분포')
plt.rc('font',family ='malgun gothic')
plt.plot(m,label = "남")
plt.plot(f,label = "여")
plt.legend()
plt.show()
```

```python
import csv
f = open('./data/gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('지역명을 입력하세요')
for row in data :
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i.replace(',','')))
        for i in row[106:]:
            f.append(int(i.replace(',',''))) 
        break           
plt.scatter(m,f,c = range(101),alpha = 0.5,cmap='jet')
# c= range(101) 은 101가지 색을 다르게 표현하겠다.
# alpha = 투명도
# cmap = 컬러맵적용
plt.colorbar()# 우측 컬러바
plt.show()
```

```python
import csv
import math
f = open('./data/gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('지역명을 입력하세요')
for row in data :
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',',''))) 
            size.append(math.sqrt(int(row[i].replace(',','')))+int(row[i+103].replace(',','')))
        break           
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=100)
plt.scatter(m,f,s=size,c = range(101),alpha = 0.5,cmap='jet')# ,s=size 점의 크기를 조절.
# c= range(101) 은 101가지 색을 다르게 표현하겠다.
# alpha = 투명도
# cmap = 컬러맵적용
plt.colorbar()# 우측 컬러바
plt.plot(range(max(m)),range(max(m)),'gray')
plt.show()
```
