```python
import matplotlib.pyplot as plt
import random
result = []
for i in range(13) :
    result.append(random.randint(1,1000))
print(sorted(result))
plt.boxplot(result)
plt.show()
```

```python
import numpy as np
result = np.array(result)
print("1/4: "+str(np.percentile(result,75)))
```

```python
import csv
f = open('./data/ta_20221201235104.csv')
data = csv.reader(f)
next(data)
dec = []
for row in data :
    month = row[1].split('-')[1]
    if row[-1] != '':
        if month =='12' :
            dec.append(float(row[-1]))
plt.boxplot(dec)
```

```python
np.percentile(dec,100)
```

```python
import csv
f = open('./data/ta_20221201235104.csv')
data = csv.reader(f)
next(data)
dec = []
aug = []
for row in data :
    month = row[1].split('-')[1]
    if row[-1] != '':
        if month =='12' :
            dec.append(float(row[-1]))
        if month =='08' :
            aug.append(float(row[-1]))


plt.boxplot(dec)
plt.boxplot(aug)
plt.show()
plt.boxplot([dec,aug])
plt.show()
```
