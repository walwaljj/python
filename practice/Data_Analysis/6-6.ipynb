```python
import matplotlib.pyplot as plt 
import csv

f = open('./data/ta_20221201235104.csv')
data = csv.reader(f)
next(data)

day = []

for i in range(31):
    day.append([])

for row in data :
    if row [-1] != '':
        if row[1].split('-')[1] == '12':
            day[int(row[1].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi = 100)
# plt.boxplot(day,showfliers=False)# 이상치(다른값들보다 값이 높거나 낮은것.)제거
plt.boxplot(day)
plt.show()
```
