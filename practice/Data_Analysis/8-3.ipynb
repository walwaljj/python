```python
import matplotlib.pyplot as plt
size = [2441,2312,1031,1233]
plt.rc('font',family="malgun gothic")
# plt.axis('equal')
label = ['a형','b형','ab형','o형']
color = ['darkmagenta','deeppink','hotpink','pink']
explode = (0,1,0.3,0.1) # 차트 돌출 효과정하기
plt.pie(size,labels=label,autopct = '%.2f%%',colors=color,explode=explode)# . n f ==> 소수점(f) n 번째 자리까지 반올림하여 출력하겠다는 의미
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./data/gender.csv')
data = csv.reader(f)
size = []
name = input('지역명을 입력하세요')
for row in data :
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m+= int(row[i+3].replace(',',''))
            f+= int(row[i+106].replace(',',''))
        break
size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt
plt.title(f'{name} 남여 인구 분포')
plt.pie(size,autopct='%.2f%%',labels=['남','여'],startangle=90,explode=(0.1,0))#startangle=90 각도 조절
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('./data/gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('지역명을 입력하세요')
for row in data :
    if name in row[0]:
        for i in row[3:104] :
            m.append(-int(i.replace(',','')))
        for i in row[106:] :
            f.append(int(i.replace(',','')))
        break

plt.title(f"{name} 인구분포")
# plt.rcParams['axes.unicode_minus'] = False
plt.rc('font',family = 'Malgun Gothic')
plt.barh(range(101),m,label = '남자')
plt.barh(range(101),f,label = '여자')
plt.legend()
plt.show()
```
