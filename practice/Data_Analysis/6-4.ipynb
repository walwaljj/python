```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('./data/ta_20221201235104.csv')
data = csv.reader(f)
next(data)

dec = []
feb = []

for row in data : 
    month = row[1].split('-')[1]
    if row[-1] != '' :
        if month == '12' :
            dec.append(float(row[-1]))
        if month == '02':
            feb.append(float(row[-1]))
```

```python
plt.hist(dec , bins = 100 , color= 'skyblue' , label = 'dec')
plt.hist(feb , bins = 100 , color= 'blue' , label = 'feb')
plt.legend()
plt.show()
```
