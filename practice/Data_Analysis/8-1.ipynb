```python
import csv
import matplotlib.pyplot as plt
f = open('./data/age.csv')
data = csv.reader(f)
result = []
name = input("이름을 입력해주세요 : ")
for row in data :
    # print(row[0])
    if name in row[0] :
        for i in row[3:] :
            result.append(int(i))
plt.style.use('ggplot')
plt.rc('font',family = 'Malgun Gothic')
plt.title(f'{name}의 인구구조')
plt.plot(result)
plt.show()
```
