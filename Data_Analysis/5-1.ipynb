```python
import csv
f = open("./data/ta_20221201235104.csv")
data = csv.reader(f)
header = next(data)
next(data)
high = []
low = []

for row in data :
    # print(row)
    if row[-1] != '' and row[-2] != '':
        date = row[1].split('-')
        if 1995 <= int(date[0]):
            if date[1] =='12' and date[2] =='14' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] =False
plt.title('   ')
plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label = 'low')
plt.legend()
plt.show()
```
