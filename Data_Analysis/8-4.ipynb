```python
import csv
f = open('./data/gender.csv')
data = csv.reader(f)
m = []
f = []

name = input(' ')
for row in data :
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break

import matplotlib.pyplot as plt
plt.title(f'{name} ')
plt.rc('font',family ='malgun gothic')
plt.plot(m,label = "")
plt.plot(f,label = "")
plt.legend()
plt.show()
```

```python
import csv
import math
f = open('./data/gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input(' ')
for row in data :
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',',''))) 
            size.append(math.sqrt(int(row[i].replace(',','')))+int(row[i+103].replace(',','')))
        break           
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=100)
plt.scatter(m,f,s=size,c = range(101),alpha = 0.5,cmap='jet')# ,s=size   .
# c= range(101)  101   .
# alpha = 
# cmap = 
plt.colorbar()#  
plt.plot(range(max(m)),range(max(m)),'gray')
plt.show()
```
